#### Data set del Torneo Apertura

<b>Torneo:</b> Nombre del torneo en curso cuando se jugó el partido <br>
<b>Fecha:</b> En qué fecha se jugó el partido. <br>
<b>Partido:</b> Número de partido dentro de la fecha. <br>
<b>Equipo_local/visitante:</b> Nombre del equipo local/visitante. <br>
<b>Goles_local/visitante: </b> Número de goles anotados por el equipo local/visitante. <br>
<b>Posesion_local/visitante:</b> Porcentaje de posesión del equipo local/visitante. <br>
<b>Tiros_arco_local/visitante: </b> Tiros al arco del equipo local/visitante. <br>
<b>Intentos_local/visitante:</b> Intentos totales del equipo local/visitante. <br>
<b>Faltas_local/visitante:</b> Faltas cometidas por el equipo local/visitante. <br>
<b>Tiros_esquina_local/visitante: </b> Tiros de esquina realizados por el equipo local/visitante. <br>
<b>Amarillas_local/visitante:</b> Tarjetas amarillas del equipo local/visitante. <br>
<b>Rojas_local/visitante: </b> Tarjetas rojas del equipo local/visitante. <br>
<b>Valor_mercado_local/visitante:</b> Valor de mercado según el equipo local/visitante. <br>
<b>Altura_media_local/visitante:</b> Altura media del equipo local/visitante. <br>
<b>Edad_media_local/visitante:</b> Edad media del equipo local/visitante. <br>
<b>Proporcion_zurdos_local/visitante:</b> Proporción de jugadores zurdos en el equipo local/visitante. <br>
<b>Resultado:</b> Resultado del encuentro. <br>
<b>Fecha_encuentro:</b> Fecha del encuentro. <br>
<b>Apuesta_local/visitante:</b> Retorno de apuesta para una victoria del equipo local/visitante. <br>
<b>Apuesta_empate:</b> Retorno de apuesta para un empate. <br>

#### Grafico con Matplot

```python
# Cantidad de Goles de local segun el torneo

plt.figure(figsize=(12, 6))  # Ajusta el tamaño de la figura
afa.groupby('torneo')['goles_local'].sum().plot(kind='bar')
plt.title('Histograma de Goles de Local por Torneo')
plt.xlabel('Torneo')
plt.ylabel('Total de Goles de Local')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para una mejor legibilidad
plt.show()
```

Se puede observar en este histograma, donde a medida que iban pasando los años decrecia la cantidad de goles de local. Y que en el 2020 no hubo torneo por pandemia.

### Seaborn

```python
plt.figure(figsize=(12, 6))  # Ajusta el tamaño de la figura
sns.barplot(x='torneo', y='goles_visitante', data=afa, estimator=sum, ci=None)
plt.title('Histograma de Goles de Local por Torneo')
plt.xlabel('Torneo')
plt.ylabel('Total de Goles de Local')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para una mejor legibilidad
plt.show()
```

Se puede apreciar a que durante el campeonato del 2016 fue el año donde mas goles hubo. Y que durante años anteriores siempre los goles de visitante fue superior a 300, a excepcion del 2016 que fue mas corto de lo normal

```python
#Metodo de pearson

#Separamos variables lineales de categoricas
variables_lineales = afa.select_dtypes(include=['int64', 'float64'])  # Incluye columnas numéricas
variables_categoricas = afa.select_dtypes(include=['object'])  # Incluye columnas de texto u otras categorías
```

```python
colormap = plt.cm.viridis
plt.figure(figsize=(12,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(variables_lineales.astype(float).corr(),linewidths=0.1,vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True)
```
